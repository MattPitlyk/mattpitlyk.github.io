### Weather Report  
This report contains minimum, average, and maximum temperatures for a week in St. Louis.

```python
# dates = ['Aug 30', 'Aug 31', 'Sep 1', 'Sep 2', 'Sep 3', 'Sep 4', 'Sep 5']
dates = [1, 2, 3, 4, 5, 6, 7]
max_temps = [84,82, 76, 80, 89, 82, 88]

ave_temps = [72.8, 73.4, 72.3, 74.9, 75.7, 71.9, 73.4]

min_temps = [60, 0, 71, 70, 69, 62, 62]
```

The Bokeh plot below is interative in this report.  
You can pan, zoom, select, and hover over data points.

```python
from bokeh.plotting import figure, output_notebook, show, ColumnDataSource
output_notebook(hide_banner=True)


source = ColumnDataSource(data=dict(
    dates=dates,
    max_temps=max_temps,
    min_temps=min_temps,
    ave_temps=ave_temps
))
TOOLTIPS = [
    ('max', '@max_temps'),
    ('ave', '@ave_temps'),
    ('min', '@min_temps')
]
p = figure(plot_width=400, plot_height=400, tooltips=TOOLTIPS,
           title="Average Temperature by Day")
p.xaxis.axis_label = 'Date in September'
p.yaxis.axis_label = 'Average Temperature'
p.circle('dates', 'ave_temps', size=10, alpha=.9, source=source)

show(p)
```
